```python
import numpy as np
import pandas as pd
import gseapy as gp
import os
```

```python
# GSEA performed on preranked gene sets, ranked by log odds ratio

# Read in the whole log odds table
raw_gene_list_rank = pd.read_csv('/Users/victoriagao/Documents/MSc/rotation/Gaiti_lab/Joan/liana__6237_2222190_C_tumor_edge.csv')

```

```python
raw_gene_list_rank
```

```python
# Filter out the L-R pairs from the same cell types
filtered_df = raw_gene_list_rank[raw_gene_list_rank['source'] != raw_gene_list_rank['target']]
filtered_df
```

```python
gene_list_w_rank = raw_gene_list_rank[['ligand.complex','receptor.complex','aggregate_rank']]
gene_list_w_rank
```

```python
# Define a list of gene sets you want to use
gene_sets = ['Reactome_2022', 'KEGG_2021_Human', 'MSigDB_Hallmark_2020']

output_folder = "/Users/victoriagao/Documents/MSc/rotation/Gaiti_lab/GSEA/results"

# Create the output folder if it doesn't exist
# os.makedirs(output_folder, exist_ok=True)
```

```python
# Split the dataframe into two: one for ligand and one for receptor
df_ligand = gene_list_w_rank[['ligand.complex', 'aggregate_rank']].copy()
df_ligand.columns = ['gene', 'aggregate_rank']

df_receptor = gene_list_w_rank[['receptor.complex', 'aggregate_rank']].copy()
df_receptor.columns = ['gene', 'aggregate_rank']

# Concatenate the dataframes
df_combined = pd.concat([df_ligand, df_receptor]).reset_index(drop=True)

print(df_combined)
```

```python
pre_res = gp.prerank(rnk=df_combined,
                    gene_sets=gene_sets,
                    threads=4,
                    min_size=1,
                    max_size=1000,
                    permutation_num=1000,
                    outdir=None,
                    seed=6,
                    verbose=True)

pre_res = pre_res.res2d.sort_values(by='NES', ascending=False)
```

```python
pre_res
```

```python
result_file_name = f"gsea_results_tumor_edge.csv"
result_file_path = os.path.join(output_folder, result_file_name)
pre_res.to_csv(result_file_path, index=False)
```
